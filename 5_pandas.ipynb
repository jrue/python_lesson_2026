```python
df["Child's Date of Entry"] = pd.to_datetime(df["Child's Date of Entry"], format="%m/%d/%y", errors="coerce")

df.set_index("Child's Date of Entry")["Length of Detention"].resample("ME").mean().plot()
```

```python
df.groupby("Child's Country of Origin")["Length of Detention"].mean()
```

```python
top10 = df.groupby("Child's Country of Origin")["Length of Detention"].mean().sort_values(ascending=False).head(10)

top10.sort_values().plot(kind='barh')
```

```python
grouped = (
    df.groupby("Child's Country of Origin")["Length of Detention"]
    .agg(mean='mean', count='count')
    .sort_values(by='mean', ascending=False)
    .head(10)
)

ax = grouped['mean'].sort_values().plot(kind='barh')

# Add labels (counts) to the right of each bar
for i, (value, count) in enumerate(zip(grouped['mean'].sort_values(), grouped['count'].loc[grouped['mean'].sort_values().index])):
    ax.text(value, i, f"{count}", va='center', ha='left')
```

```python
top_countries = df["Child's Country of Origin"].value_counts().head(10).index
df_top = df[df["Child's Country of Origin"].isin(top_countries)]
order = df_top.groupby("Child's Country of Origin")["Length of Detention"].median().sort_values(ascending=False).index

sns.boxplot(
    data=df_top,
    y="Child's Country of Origin",
    x="Length of Detention",
    order=order,
    showfliers=False
)
```

```python
sns.boxplot(data=df, y="Child's Gender", x="Length of Detention", showfliers=False)
```

```python
pivot = (
    df[df["Child's Country of Origin"].isin(top_countries)]
    .pivot_table(
        values="Length of Detention",
        index="Child's Country of Origin",
        columns="Sponsor Category",
        aggfunc='median'
    )
)

import seaborn as sns
sns.heatmap(pivot, cmap="coolwarm")
```
